### Lines and Circles
- The following graphics primitives are supported:

    - arrow() - an arrow from a min point to a max point.
    - circle() - a circle with given radius
    - ellipse() - an ellipse with given radii and angle
    - arc() - an arc of a circle or an ellipse
    - disk() - a filled disk (i.e. a sector or wedge of a circle)
    - line() - a line determined by a sequence of points (this need not be straight!)
    - point() - a point
    - text() - some text
    - polygon() - a filled polygon


```python
circle((1,0), 1)
```

```python
line([(1,2),(2,3)])
```

```python
circle((0,0), 2) +line([(0,0),(1.42,1.42)])+ arrow((0,0), (-2,1)) + point((1,0.5))
```

```python
f(x) = (x-3)*(x-5)*(x-7)+40
P = line([(2,0),(2,f(2))], color='black')
P += line([(8,0),(8,f(8))], color='black')
P += polygon([(2,0),(2,f(2))] + [(x, f(x)) for x in [2,2.1,..,8]] + [(8,0),(2,0)],  rgbcolor=(0.8,0.8,0.8),aspect_ratio='automatic')
P += text("$\\int_{a}^b f(x) dx$", (5, 20), fontsize=16, color='black')
P += plot(f, (1, 8.5), thickness=3)
P    # show the result
```

### Change Plot Properties
- We can change the figure propreties as follows
- color - (Default: blue)
- linestyle - (default: -) The style of the line, which is one of

    "-" or "solid"
    "--" or "dashed"
    "-." or "dash dot"
    ":" or "dotted"
- marker - The style of the markers, which is one of

    "None" or " " or "" (nothing)  default
    "," (pixel), "." (point)
    "_" (horizontal line), "|" (vertical line)
    "o" (circle), "p" (pentagon), "s" (square), "x" (x), "+" (plus), "*" (star)
    "D" (diamond), "d" (thin diamond)

```python
 plot(sin(x), x, -2*pi, 2*pi, color = 'red')
```

```python
 plot(sin(x), x, -2*pi, 2*pi, linestyle='--')
```

```python
 plot(sin(x), x, -2*pi, 2*pi, linestyle='--', thickness = 3)
```

```python
plot(sin(x), x, -2*pi, 2*pi, color = 'purple', \
axes_labels =[ 'x','sin(x)'])
```

```python
plot(x, (x,-8,8), ticks=[3,4])
```

```python
plot(sin(x), x, -2*pi, 2*pi, \
color = 'purple', \
axes_labels =[ 'x','sin(x)'], \
ticks = pi/4, \
tick_formatter=pi)
```

### Combine Plots together
- To add more than one plot

```python
p = plot(sin(x), x, -2*pi, 2*pi, \
color = 'blue', \
axes_labels =[ 'x','sin(x)'], \
ticks = 1, \
tick_formatter=2)
q = plot(sin(6*x), x, -2*pi, 2*pi, \
color = 'red', \
linestyle = 'dotted')
show(p+q)
```

```python
p = plot(sin(x), x, -2*pi, 2*pi, color = 'blue', \
axes_labels =[ 'x','sin(x)'], \
ticks = pi/3, tick_formatter=pi, \
legend_label = 'sin(x)' )
q = plot(sin(6*x), x, -2*pi, 2*pi, color = 'red', \
linestyle = 'dotted', \
legend_label = '$\sin(6 \cdot x)$' )
show(p+q)
```

- We can use the function **legend_label** to control the legends of functions

```python
f(x) = 1+x/2+x^2/2+x^3/6+x^4/24
g(x) = e^x
plot(f(x), x, -1,1, legend_label = "$"+str(f(x))+"$") + \
plot(g(x), x, -1,1, legend_label = "$"+str(g(x))+"$")
```

- We can have multiple plots as well

```python
g1 = plot(sin(x), (x, -4*pi, 4*pi))
g2 = plot(sin(6*x), (x, -4*pi, 4*pi))
g3 = plot(sin(12*x), (x, -4*pi, 4*pi))
L = plot(x,x, -1,1)
show(graphics_array( \
[[g1, g2,L], \
[L ,L, g3]], 2, 3))
```

```python
#define the function
f(x)=e^x
#generate the graph of the function
graph = plot(f(x),x,-0.5,3.5)
#generate the secant line
secant = line([(1,f(1)),(3,f(3))], color='black')
#add text to label these points
text1 = text("x=1", (1, f(1)+4), fontsize=20, color='red')
text2 = text("x=3", (3, 21), fontsize=20, color='red')
#display all in one
(graph+secant+text1+text2).show()
```

- Plots of Functions with Asymptotes:

```python
empty = Graphics()    
empty.axes(False)
p2 = plot(1/(x^3-x), -2, 2, ymin = -5, ymax = 5)
p3 = plot(1/(x^3-x), (x,-2, 2), detect_poles=True, ymin = -5, ymax = 5)
p4 = plot(1/(x^3-x), (x,-2, 2), detect_poles='show',ymin = -5, ymax = 5)
p=graphics_array([[p2, empty],[p3, p4]], 2, 2)
p.show(ymin=-5, ymax=5)
```

### Parametric Plot
- Sage Math can also plot parametric functions. For example,

```python
var('t')
parametric_plot( (cos(t), sin(t)), (t, 0, 2*pi))
```

```python
parametric_plot( (cos(x), sin(x)), (x, 0, 6*pi),color='blue',fill=True, fillcolor='red')
```

```python
parametric_plot( (cos(x), sin(x)*cos(x)), (x, 0, 6*pi),color='blue',fill=True, fillcolor='red')
```

```python
parametric_plot([cos(x) + 2 * cos(x/4), sin(x) - 2 * sin(x/4)], (x,0, 8*pi), fill=True)
```

```python
f(x) = sin(x)
g(x) = sin(2*x)
h(x) = sin(4*x)
p1 = plot(f,(-2*pi,2*pi),color=hue(0.5)) # long time
p2 = plot(g,(-2*pi,2*pi),color=hue(0.9)) # long time
p3 = parametric_plot((f,g),(0,2*pi),color=hue(0.6)) # long time
p4 = parametric_plot((f,h),(0,2*pi),color=hue(1.0)) # long time
graphics_array(((p1,p2),(p3,p4)))
```
